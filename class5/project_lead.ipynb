```python
newDF = pd.DataFrame()

# df[df['Negócio - Won time'].notnull()] - remove null
# df.filter(like= ' ', axis=0) - filter example
```

```python
# Pegar o valor - Negócio - Value
negocioValue = df['Negócio - Value'].tolist()

def getMoney(v):
    return int(v.split()[0].replace(',', '').replace('.', ''))
def getCurrency(v):
    return v.split()[1]

negocioValueMoney = [*map(getMoney, negocioValue)]
negocioValueCurrency = [*map(getCurrency, negocioValue)]
```

```python
# Insere moeada e valor
# newDF = newDF.drop(['Negócio - Valor'], axis=1)
# newDF = newDF.drop(['Negócio - Moeda'], axis=1)
newDF.insert(0, 'Negócio - Valor', negocioValueMoney)
newDF.insert(1, 'Negócio - Moeda', negocioValueCurrency)
```

```python
# Adiciona estágio do negócio
newDF.insert(2, 'Negócio - Estágio', df['Negócio - Stage'].tolist())
```
