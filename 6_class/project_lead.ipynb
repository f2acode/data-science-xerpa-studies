<h2><a id="leitura" class="anchor">1 - Leitura e Entendimento do Arquivo</a></h2>

```python
# Bibliotecas:
import pandas as pd

# Configurações Jupyter Notebooks
pd.set_option('display.max_columns', 300)
pd.set_option('display.max_rows', 300)

# Ler arquivo Deals
df_deals = pd.read_csv('arquivos/deals.csv', sep=';')

# Shape arquivo Deals
print("Número de linhas: " + str(df_deals.shape[0]))
print("Número de Colunas: " + str(df_deals.shape[1]))
```

```python
# Verificar arquivo:
df_deals.head()
```

```python
# Ajuste nome das variáveis:
import re
from unicodedata import normalize

for column in df_deals.columns:
    newname = column.split(sep='-')[1]
    newname = newname.strip() # eliminar espaços em branco no ínicio e final
    newname = newname.lower() # converter todas as letras para minúscula
    newname = re.sub(r'[^\w\s]','',newname) # eliminar pontuação
    newname = re.sub(' ','_',newname) # substituir espaço por underscore
    newname = normalize('NFKD', newname).encode('ASCII', 'ignore').decode('ASCII') # eliminar acentos
    df_deals.rename(columns={column:newname},inplace=True)

df_deals.head()
```

```python
# Nome das colunas
df_deals.columns
```

```python
df_deals.dtypes
```

```python
# Ajuste campo value

def retorna_valor(x):
    value = x.split(sep=' ')[0]
    value = re.sub(r'[^\w\s]','',value)
    return(value)

def retorna_moeda(x):
    return(x.split(sep=' ')[1])

df_deals['value_1'] = df_deals.value.apply(lambda x: retorna_valor(x)).astype('float64')/100
df_deals['value_moeda'] = df_deals.value.apply(lambda x: retorna_moeda(x))

df_deals['value'] = df_deals['value_1']

df_deals.head(300)
```

```python
# Verificar as colunas criadas
print(df_deals['value_1'].describe())
print('\n')
print(df_deals.value_moeda.unique())
```

```python
# Excluir variáveis não desnecessárias:
df_deals['value'] = df_deals['value_1']
df_deals.drop(['value_1','value_moeda'],axis=1,inplace=True)
df_deals.head()
```

```python
df_deals.dtypes
```

```python
# Criar campos derivados:
def derivateDate (column):

    #Derivação campos de datas
    df_deals[column + '_date'] = df_deals[column].astype('datetime64[D]')
    #deal_created_year
    df_deals[column + '_year'] = df_deals[column + '_date'].apply(lambda x: x.year)
    #deal_created_month
    df_deals[column + '_month'] = df_deals[column + '_date'].apply(lambda x: x.month)
    #deal_created_day
    df_deals[column + '_day'] = df_deals[column + '_date'].apply(lambda x: x.day)
    #deal_created_quarter
    df_deals[column + '_quarter'] = df_deals[column + '_date'].apply(lambda x: x.quarter)
    #deal_created_weekday
    df_deals[column + '_weekday'] = df_deals[column + '_date'].apply(lambda x: x.dayofweek)
    #drop column
    df_deals.drop([column],axis=1,inplace=True)
```

```python
# Fazer o mesmo para os demais campos de data:
derivateDate('deal_created')
derivateDate('lost_time')
derivateDate('last_activity_date')
derivateDate('last_email_received')
derivateDate('last_email_sent')
derivateDate('last_stage_change')
derivateDate('expected_close_date')
derivateDate('deal_closed_on')
derivateDate('next_activity_date')
derivateDate('update_time')
# df_deals.filter(regex='\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2}', axis=1)
#df_deals.drop(['deal_created_date', 'deal_created_d', 'deal_created_h', ''],axis=1,inplace=True)
```
