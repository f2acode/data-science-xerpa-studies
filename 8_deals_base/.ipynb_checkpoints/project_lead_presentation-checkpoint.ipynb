<h3><a id="column_names" class="anchor">2.2 - Ajuste do nome das colunas</a></h3>

```python
import re
from unicodedata import normalize

for column in df_deals.columns:
    newname = column.split(sep='-')[1]
    newname = newname.strip() # eliminar espaços em branco no ínicio e final
    newname = newname.lower() # converter todas as letras para minúscula
    newname = re.sub(r'[^\w\s]','',newname) # eliminar pontuação
    newname = re.sub(' ','_',newname) # substituir espaço por underscore
    newname = normalize('NFKD', newname).encode('ASCII', 'ignore').decode('ASCII') # eliminar acentos
    df_deals.rename(columns={column:newname},inplace=True)
```

```python
import numpy as np

def redefineTypes():
    types = []
    for c in df_deals.columns:
        if df_deals[c].dtype == object:
            types.append('Text')
        elif df_deals[c].dtype == 'float64':
            types.append('Decimal')
        elif df_deals[c].dtype == 'int64':
            types.append('Integer')
        else:
            types.append('NOOOONE')
    return types
    
def getTotalFilled():
    totalFilled = []
    for c in df_deals.columns:
        totalFilled.append(f'{df_deals[c].count()} / {df_deals.shape[0]}')
    return totalFilled
    
dataTypes = pd.Series(redefineTypes(), index=df_deals.columns)
totalFilled = pd.Series(getTotalFilled(), index=df_deals.columns)

df_details = pd.DataFrame()
df_details.insert(0, "tipo_de_dado", dataTypes) 
df_details.insert(1, "total_preenchido", totalFilled) 

df_details
```

```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

completeness = []
completenessDict = ({})

for i in df_deals.columns:
    completeness.append([df_deals[i].count()])
    completenessDict[i] = df_deals[i].count()
    
plt.gcf().set_size_inches(11.7, 10.27)

sns.barplot(data=completeness)
```

Variáveis removidas por outros motivos:

<h3><a id="value_field" class="anchor">2.2 - Ajuste do campo valor</a></h3>

Segregar o campo value em tipo da moeda e valor, desta forma conseguimos extrair alguns valores interessantes.

* Média
* Valor total

```python
def retorna_valor(x):
    value = x.split(sep=' ')[0]
    value = re.sub(r'[^\w\s]','',value)
    return(value)

def retorna_moeda(x):
    return(x.split(sep=' ')[1])

df_deals['value'] = df_deals.value.apply(lambda x: retorna_valor(x)).astype('float64')/100
df_deals['weighted_value'] = df_deals.weighted_value.apply(lambda x: retorna_valor(x)).astype('float64')/100
df_deals.head(300)

# df_deals['value'].describe()
```

<h3><a id="field_cases" class="anchor">2.3 - Ajuste termos semelhantes</a></h3>

É possível verificar a presença de termos semelhantes, alguns exemplos:
* Palavras escritas começando com maiúsculo ou minúsculo;
* Palavras com acento e sem acento;
* Palavras com espaço no fim, início etc.

```python
def print_details(column_name):
    '''print some details of columns with object type'''
    df_temp = df_deals.loc[:,[column_name,'currency']].groupby(column_name).count().reset_index()
    df_temp = df_temp.rename(columns={'currency':'qnt'}).sort_values('qnt',ascending=False)
    print("Quantidade de categorias: ", df_temp.shape[0])
    # print(f'{df_temp} \n')
```

```python
import re
from unicodedata import normalize

df_deals = pd.read_csv('arquivos/deals_terms_adjustment.csv', sep=';')
df_deals_object = df_deals.select_dtypes('object')

def filterChar(char):
    if '*' not in char or '+' not in char or '?' not in char:
        return False
    return True

def normalizeStrings(v):
    if not pd.isnull(v):
        newV = v.strip()
        newV = newV.lower()
        newV = re.sub(r'[^\w\s]','',newV)
        newV = re.sub(' ','_',newV)
        newV = normalize('NFKD', newV).encode('ASCII', 'ignore').decode('ASCII')
        return newV
    else:
        return v
    
for c in df_deals_object.columns:
    print(c)
    if c != 'deal_created' and c != 'lost_time' and c != 'last_activity_date' and c != 'last_stage_change' and c != 'expected_close_date' and c != 'deal_closed_on' and c != 'update_time':
        print_details(c)
        df_deals[c] = df_deals_object[c].apply(normalizeStrings)
        print_details(c)
    
df_deals.head()
```

<h2><a id="derivate_fields" class="anchor">3 - Derivação de campos</a></h2>

<h3><a id="value_field" class="anchor">3.1 - Derivação de datas</a></h3>

Importante para conseguir entender a informação de forma segregada, ou seja, semanalmente, mensalmente, semestralmente etc.

Resumidamente: a derivação de variáveis pode trazer novos dados, ou uma maneira diferente de olhar para os mesmos dados.

```python
def derivateDate (column):
    df_deals[column + '_date'] = df_deals[column].astype('datetime64[D]')
    df_deals[column + '_year'] = df_deals[column + '_date'].apply(lambda x: x.year)
    df_deals[column + '_month'] = df_deals[column + '_date'].apply(lambda x: x.month)
    df_deals[column + '_day'] = df_deals[column + '_date'].apply(lambda x: x.day)
    df_deals[column + '_quarter'] = df_deals[column + '_date'].apply(lambda x: x.quarter)
    df_deals[column + '_weekday'] = df_deals[column + '_date'].apply(lambda x: x.dayofweek)
    df_deals.drop([column],axis=1,inplace=True)
    
derivateDate('deal_created')
derivateDate('lost_time')
derivateDate('last_activity_date')
# derivateDate('last_email_received')
# derivateDate('last_email_sent')
derivateDate('last_stage_change')
derivateDate('expected_close_date')
derivateDate('deal_closed_on')
# derivateDate('next_activity_date')
derivateDate('update_time')

df_deals.head() 
```

A distribuição mostra como os valores geralmente se comportam dentro da amplitude disponível.

Os outliers são valores que estão fora do comum em relação ao conjunto de dados disponíveis e que poderiam prejudicar a análise dos dados.

Outras variáveis checadas:

* update_time_month
* activities_to_do
* done_activities

```python
df_deals = pd.read_csv('arquivos/deals_derivations_adjustment.csv', sep=';')

import numpy as np

df_deals_num = df_deals.select_dtypes(['float64','int64'])

# Distribuição:
bins = 10
var = 'deal_created_month'

bin_values = np.arange(start=df_deals_num[var].min(), 
                       stop=df_deals_num[var].max(), 
                       step=((df_deals_num[var].max()-df_deals_num[var].min())/bins))
```
