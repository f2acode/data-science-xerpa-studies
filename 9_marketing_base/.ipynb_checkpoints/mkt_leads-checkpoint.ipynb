<h3><a id="column_names" class="anchor">2.2 - Ajuste do nome das colunas</a></h3>

```python
import re
from unicodedata import normalize

for column in df_leads_mkt.columns:
    newname = column
    #if newname[0] == '#':
    #    newname = newname.replace('#', '', 1)
    newname = newname.strip()
    newname = newname.lower()
    newname = re.sub(r'[^\w\s]','',newname) # eliminar pontuação
    newname = re.sub(' ','_',newname) # substituir espaço por underscore
    newname = normalize('NFKD', newname).encode('ASCII', 'ignore').decode('ASCII') # eliminar acentos
    df_leads_mkt.rename(columns={column:newname},inplace=True)
```

```python
import numpy as np

def redefineTypes():
    types = []
    for c in df_leads_mkt.columns:
        if df_leads_mkt[c].dtype == object:
            types.append('Text')
        elif df_leads_mkt[c].dtype == 'float64':
            types.append('Decimal')
        elif df_leads_mkt[c].dtype == 'int64':
            types.append('Integer')
        else:
            types.append('NOOOONE')
    return types
    
def getTotalFilled():
    totalFilled = []
    for c in df_leads_mkt.columns:
        totalFilled.append(f'{df_leads_mkt[c].count()} / {df_leads_mkt.shape[0]}')
    return totalFilled
    
dataTypes = pd.Series(redefineTypes(), index=df_leads_mkt.columns)
totalFilled = pd.Series(getTotalFilled(), index=df_leads_mkt.columns)

df_details = pd.DataFrame()
df_details.insert(0, "tipo_de_dado", dataTypes) 
df_details.insert(1, "total_preenchido", totalFilled) 

df_details
```

```python
import re
from unicodedata import normalize

df_leads_mkt = pd.read_csv('arquivos/leads_mkt_terms_adjustment.csv', sep=';')
df_leads_mkt_object = df_leads_mkt.select_dtypes('object')

def filterChar(char):
    if '*' not in char or '+' not in char or '?' not in char:
        return False
    return True

def normalizeStrings(v):
    if not pd.isnull(v):
        newV = v.strip()
        newV = newV.lower()
        newV = re.sub(r'[^\w\s]','',newV)
        newV = re.sub(' ','_',newV)
        newV = normalize('NFKD', newV).encode('ASCII', 'ignore').decode('ASCII')
        return newV
    else:
        return v
    
for c in df_leads_mkt_object.columns:
    print(c)
    if c != 'data_da_primeira_conversao' and c != 'data_da_ultima_conversao' and c != 'email' and c != 'tags' and c != 'url_publica' and c != 'origem_da_ultima_conversao':
        print_details(c)
        df_leads_mkt[c] = df_leads_mkt_object[c].apply(normalizeStrings)
        print_details(c)
    
df_leads_mkt.head()
```

<h2><a id="derivate_fields" class="anchor">3 - Derivação de campos</a></h2>

<h3><a id="value_field" class="anchor">3.1 - Derivação de datas</a></h3>

Importante para conseguir entender a informação de forma segregada, ou seja, semanalmente, mensalmente, semestralmente etc.

Resumidamente: a derivação de variáveis pode trazer novos dados, ou uma maneira diferente de olhar para os mesmos dados.

```python
def derivateDate (column):
    df_leads_mkt[column + '_date'] = df_leads_mkt[column].astype('datetime64[D]')
    df_leads_mkt[column + '_year'] = df_leads_mkt[column + '_date'].apply(lambda x: x.year)
    df_leads_mkt[column + '_month'] = df_leads_mkt[column + '_date'].apply(lambda x: x.month)
    df_leads_mkt[column + '_day'] = df_leads_mkt[column + '_date'].apply(lambda x: x.day)
    df_leads_mkt[column + '_quarter'] = df_leads_mkt[column + '_date'].apply(lambda x: x.quarter)
    df_leads_mkt[column + '_weekday'] = df_leads_mkt[column + '_date'].apply(lambda x: x.dayofweek)
    df_leads_mkt.drop([column],axis=1,inplace=True)
    
derivateDate('data_da_primeira_conversao')
derivateDate('data_da_ultima_conversao')

df_leads_mkt.head() 
```

<h2><a id="outliers" class="anchor">4 - Entendendo a distribuição e Eliminação de outliers</a></h2>

A distribuição mostra como os valores geralmente se comportam dentro da amplitude disponível.

Os outliers são valores que estão fora do comum em relação ao conjunto de dados disponíveis e que poderiam prejudicar a análise dos dados.

```python
df_leads_mkt = pd.read_csv('arquivos/leads_mkt_derivations_adjustment.csv', sep=';')

import numpy as np

df_leads_mkt_num = df_leads_mkt.select_dtypes(['float64','int64'])

# Distribuição:
bins = 10
var = 'conversoes'

bin_values = np.arange(start=df_leads_mkt_num[var].min(), 
                       stop=df_leads_mkt_num[var].max(), 
                       step=((df_leads_mkt_num[var].max()-df_leads_mkt_num[var].min())/bins))
```

```python
df_leads_mkt_num[var].hist(bins=bin_values, figsize=[14,6])
```
